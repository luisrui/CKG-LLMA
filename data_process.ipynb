```python
import pandas as pd
rateinfo = pd.read_csv('dataset/AmazonBook/Books_rating.csv')
```

```python
rateinfo.shape
```

```python
### Filter for positive reviews(Changing the logic for filtering the reviews)
sep = 4
rateinfo_filtered = rateinfo.dropna(subset=['User_id'])
filtered_reviews = rateinfo_filtered[rateinfo_filtered['review/score'] >= sep]
```

```python
filtered_reviews.shape
```

```python
filtered_reviews.head()
```

```python
columns = ['User_id', 'Id', 'review/text', 'review/summary']
filtered_reviews_save = filtered_reviews[columns]
filtered_reviews_save = pd.DataFrame(filtered_reviews_save, columns=columns)
filtered_reviews_save.to_csv(f'./dataset/AmazonBook/data_all.csv', index=False)
filtered_reviews_save.head()
```
